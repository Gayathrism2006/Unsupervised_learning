```python
from google.colab import files
uploaded=files.upload()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler

```

```python
import hdbscan
df=pd.read_csv('mall.csv')
df.drop(['CustomerID'],axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df.head()
```

```python
x=df
scaler=MinMaxScaler()
v=scaler.fit_transform(x)
```

```python
model=hdbscan.HDBSCAN(min_cluster_size=12,metric='euclidean',cluster_selection_method='eom',min_samples=5)
y_predict=model.fit_predict(v)
y_predict
```

```python
score=silhouette_score(v,y_predict)
score
```

```python
plt.scatter(v[:,2],v[:,3],c=y_predict)
plt.title('hdbscan')
plt.xlabel('annual')
plt.ylabel('spending')
plt.show()
```

```python
model.condensed_tree_.plot()
plt.show()
```
